## 染色体大小

```python
file_path="/cluster/home/futing/Project/GBM/Corigami/corigami_data/data/hg38/gbm/genomic_features/hg38.chrom.sizes"
data_dict = {}

with open(file_path, 'r') as file:
    lines = file.readlines()

for line in lines:
    key, value = line.strip().split('\t')
    data_dict[key] = int(value)
    
top_24_keys = list(data_dict.keys())[:24]

new_dict = {}
for key in top_24_keys:
    new_dict[key] = data_dict[key]
print(new_dict)
```

## 创造空文件

```python
#chrom_sizes = {"chr1": 1000000, "chr2": 1500000, "chr3": 800000}
chrom_sizes = new_dict
output_file = "zero_values.bw"

bw = pyBigWig.open(output_file, "w")

chromosomes = [(k, v) for k, v in chrom_sizes.items()]

# 添加文件头信息
bw.addHeader(chromosomes, maxZooms=0)
# 写入0值数据
for chrom, size in chrom_sizes.items():
    # 初始化一个值全为0的数组
    values = [0.0] * int(size/500)
    # 写入0值数据
    bw.addEntries(chrom, 0, values=values, span=500,step= int(size/500))
    print(chrom,size,len(values))

# 关闭文件
bw.close()

print("Zero-value BigWig file created:", output_file)
```
