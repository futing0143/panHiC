```python
import os.path as op
import matplotlib.pyplot as plt
import numpy as np
import pandas
import h5py
import os

import cooler
```

```python
NPC = cooler.Cooler('otherGBM/mcoolfile/1M/NPC_1M.cool')
```

```python
print(f'chromosomes: {NPC.chromnames}, binsize: {NPC.binsize}')
```

```python
chromstarts = []
for i in NPC.chromnames:
    print(f'{i} : {NPC.extent(i)}')
    chromstarts.append(NPC.extent(i)[0])
```

```python
f, ax = plt.subplots(
    figsize=(7,6))
im = ax.matshow((NPC.matrix(balance=False)[:]),vmax=500);
plt.colorbar(im ,fraction=0.046, pad=0.04, label='raw counts')
ax.set(xticks=chromstarts, xticklabels=NPC.chromnames,
       xlabel='position, chrom#', ylabel='position, bin#')
ax.xaxis.set_label_position('top')
```

```python
from matplotlib.ticker import EngFormatter
bp_formatter = EngFormatter('b')

def format_ticks(ax, x=True, y=True, rotate=True):
    if y:
        ax.yaxis.set_major_formatter(bp_formatter)
    if x:
        ax.xaxis.set_major_formatter(bp_formatter)
        ax.xaxis.tick_bottom()
    if rotate:
        ax.tick_params(axis='x',rotation=45)

f, axs = plt.subplots(
    figsize=(14,4),
    ncols=3)

ax = axs[0]
im = ax.matshow(NPC.matrix(balance=False)[:], vmax=2500);
plt.colorbar(im, ax=ax ,fraction=0.046, pad=0.04, label='raw counts');
ax.set_xticks(chromstarts)
ax.set_xticklabels(NPC.chromnames)
ax.set_yticks(chromstarts)
ax.set_yticklabels(NPC.chromnames)
ax.xaxis.tick_bottom()
ax.set_title('All data')

ax = axs[1]
im = ax.matshow(
    NPC.matrix(balance=False).fetch('chr8'),
    vmax=2500,
    extent=(0,NPC.chromsizes['chr8'], NPC.chromsizes['chr8'], 0)
);
plt.colorbar(im, ax=ax ,fraction=0.046, pad=0.04, label='raw counts');
ax.set_title('chr8', y=1.08)
ax.set_ylabel('position, Mb')
format_ticks(ax)

ax = axs[2]
start, end = 30_000_000, 60_000_000
region = ('chr8', start, end)
im = ax.matshow(
    NPC.matrix(balance=False).fetch(region),
    vmax=2500,
    extent=(start, end, end, start)
);
ax.set_title(f'chr8:{start:,}-{end:,}', y=1.08)
plt.colorbar(im, ax=ax ,fraction=0.046, pad=0.04, label='raw counts');
format_ticks(ax)
plt.tight_layout()
```
