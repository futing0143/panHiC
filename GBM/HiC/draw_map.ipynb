```python
import os
import cooler
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

# 定义格式刻度的函数
def format_ticks(ax, labels, labelsize=8): 
    ax.xaxis.set_ticks_position('bottom')  # 确保x轴刻度在底部
    ax.set_xticks(range(len(labels)))  # 设置x轴的刻度位置
    ax.set_xticklabels(labels, rotation=45, fontsize=labelsize)
    ax.set_yticklabels([])  # 隐藏y轴刻度标签
    ax.tick_params(axis='y', which='both', length=0)  # 隐藏y轴刻度线

# 定义基因区域
gene_regions = {
    'SOX2': ('chr3', 181000000, 185000000),
    'POU3F2': ('chr6', 98000000, 103000000),
    'SALL2': ('chr14', 20000000, 25000000),
    'OLIG2': ('chr21', 30000000, 35000000)
}

# 文件夹路径和输出PDF文件名
cooler_folder_path = '/cluster/home/tmp/GBM/HiC/02data/03cool/1000000/norm'
output_pdf_filename = '/cluster/home/jialu/GBM/4G_15cooler_1Mmap.pdf'

# 指定的文件名顺序
file_order = ['ts543', 'ts667', 'GB567', 'GB583', 'G523', 'A172', 'U87', 'U343', 'U118', 'SW1088', 'GB180', 'GB182', 'GB183', 'GB238', 'GB176']

# 根据指定顺序筛选文件列表
file_list = [os.path.join(cooler_folder_path, f"{file}_normalized.cool") for file in file_order]

# 创建一个PDF页面对象
pp = PdfPages(output_pdf_filename)

# 对每个基因区域绘制15个子图
for gene_name, (chr_name, start, end) in gene_regions.items():
    fig, axs = plt.subplots(3, 5, figsize=(25, 15))  # 3行5列的子图布局

    for i, file_basename in enumerate(file_order):
        file_path = file_list[i]  # 获取当前文件的完整路径
        ax = axs[i // 5, i % 5]  # 获取当前子图的轴对象
        
        # 加载cool文件
        c = cooler.Cooler(file_path)
        
        # 定义要绘制的染色体区域
        region = (chr_name, start, end)
        
        # 从cooler对象中获取矩阵数据并绘制
        im = ax.matshow(
            c.matrix(balance=False).fetch(region),
            vmax=2500,
            aspect='auto'
        )
        
        # 设置子图标题
        ax.set_title(f'{file_basename}', fontsize=10)
        
        region_length = (end - start) // 4  # 将区域分为5个等分
        tick_positions = range(start, end, region_length)  # 生成刻度位置列表
        tick_labels = [f'{pos}' for pos in tick_positions]  # 生成刻度标签列表
        
        # 调用修改后的format_ticks函数
        format_ticks(ax, tick_labels, labelsize=8)
    # 调整子图间距
    plt.subplots_adjust(hspace=0.4, wspace=0.4)
    
    # 设置整体图形的标题
    plt.suptitle(f'Around {gene_name} : {chr_name}:{start}-{end}', fontsize=16)
    # 将图形保存到PDF
    pp.savefig(fig, bbox_inches='tight')
    
    # 清空当前图形，为下一张图做准备
    plt.close()

# 关闭PDF页面对象
pp.close()

print(f'All Hi-C plots have been saved to {output_pdf_filename}')
```
